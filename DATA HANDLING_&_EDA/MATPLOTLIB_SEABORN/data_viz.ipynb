IMPORTS & SETUP

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="Set2")
```

LOAD DATASET & DATA OVERVIEW

```python
df = sns.load_dataset("tips")
```

UNIVARIATE ANALYSIS

```python
# Numerical Distribution (Histogram)
sns.histplot(data=df, x='total_bill', kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
# Outlier Detection (Boxplot)
sns.boxplot(x=df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.show()
```

BIVARIATE ANALYSIS

```python
# Scatter Plot
sns.scatterplot(data=df, x="total_bill", y="tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Scatter Plot with Hue
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs Tip by Sex")
plt.show()
```

CATEGORICAL ANALYSIS

```python
# Count Plot
sns.countplot(data=df, x="day")
plt.title("Visits per Day")
plt.show()
```

```python
# Count PLot with Hue
sns.countplot(data=df, x="day", hue="time")
plt.title("Visits per Day by Time")
plt.show()
```

```python
# Bar Plot
sns.barplot(data=df, x="day", y="total_bill", estimator="mean", errorbar=None)
plt.title("Average Total Bill per Day")
plt.show()
```

```python
# Bar Plot by Hue
sns.barplot(data=df, x="day", y="total_bill", hue="time", estimator="mean", errorbar=None)
plt.title("Average Total Bill per Day by Time")
plt.show()
```

DISTRIBUTION ACROSS CATEGORIES

```python
# Boxplot
sns.boxplot(data=df, x="day", y="total_bill")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
# Violin Plot
sns.violinplot(data=df, x="day", y="total_bill")
plt.title("Distribution Shape by Day")
plt.show()
```

MULTIVARIATE ANALYSIS

```python
# Pair Plot
sns.pairplot(df, hue="sex") 
plt.suptitle("Pair Plot of Tips Dataset", y=1.02)
plt.show()
```

TREND & RELATIONSHIP STRENGTH

```python
# Regression Plot
sns.regplot(data=df, x="total_bill", y="tip")
plt.title("Regression of Tip on Total Bill")    
plt.show()
```

```python
# Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
